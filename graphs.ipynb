```python
### Initialization
import pandas as pd  
# import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

floor = pd.read_csv("libFloor.csv")
```

```python
quarters = floor['Quarter'].unique().tolist()
quarters.remove('Spring 2023')
print(quarters)
```

```python
new_floor = floor[~(floor['Location'] == 'UCSB Library')]

new_floor['Timestamp'] = pd.to_datetime(new_floor['Timestamp'])

def shorten(timestamp):
    return timestamp.strftime("%Y-%m-%d")

new_floor['Timestamp'] = new_floor['Timestamp'].apply(shorten)
new_floor.head()
# new_floor_f21 = new_floor[new_floor['Quarter'] == 'Fall 2021']
```

```python
fall22 = new_floor[new_floor['Quarter'] == 'Fall 2022']
winter22 = new_floor[new_floor['Quarter'] == 'Winter 2022']
spring22 = new_floor[new_floor['Quarter'] == 'Spring 2022']
```

```python
def floor_chart(qtr):
    sns.set_style('darkgrid')
    sns.set(rc={'figure.figsize':(14,8)})

    floor_qtr = new_floor[new_floor['Quarter'] == qtr]
    ax = sns.lineplot(data=floor_qtr, x ='Timestamp', y = 'Average Occupancy',
                        hue='Location', palette='viridis',
                        legend='full', lw=2, ci=None)

    # reassign/group tick labels 
    # stagger or spread out date time ticks 
    ax.xaxis.set_major_locator(ticker.MultipleLocator(16))
    plt.legend(bbox_to_anchor=(1, 1))
    plt.ylabel('Average Occupancy')
    plt.xlabel('Timestamp')
    plt.title(qtr)

    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
    plt.show()
```

```python
for qtr in quarters: 
    floor_chart(qtr)
# look into order reversal errors 
```
